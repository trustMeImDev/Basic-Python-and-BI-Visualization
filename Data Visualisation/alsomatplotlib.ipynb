```python
#step-02--->import the dataset
dataset = pd.read_csv('F1_2022_data.csv')
print(dataset)
```

```python

plt.hist(dataset['Points'],color='red',edgecolor='white',bins=6)
plt.title('Race Points')
plt.xlabel('Points')
plt.ylabel('Frequency')
plt.show()
```

```python
#scatter plot demo
import matplotlib.pyplot as plt
plt.scatter(dataset['Wins'], dataset['Constructor'],c='red')
plt.title("Wins of each Constructor Team")
plt.xlabel('Wins')
plt.ylabel('Constructor')
plt.show()
```

```python
#demo of bar plot
counts=[454,308,305]
FuelType=('VER','LEC','PER')
index=np.arange(len(FuelType))
plt.bar(index,counts,color=['blue','red','blue'])
plt.title('bar plot of top points driver')
plt.xlabel('Driver Code')
plt.ylabel('Points')
plt.xticks(index,FuelType, rotation=90)
plt.show()
```

```python
import seaborn as sns
corel=dataset.corr()
sns.heatmap(corel, vmax=.6, square=True)
sns.heatmap(corel, vmax=.6, annot=True,square=True)
plt.show()

#ERROR: Heatmap not compatible with dataset due to datatype of the attributes!
```
