```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
image_path = 'image.jpg'
image = Image.open(image_path)
```

```python
rgb = np.array(image)
plt.title("Original Image")
plt.imshow(rgb)
plt.show()
```

```python
red = rgb[:, :, 0]
green = rgb[:, :, 1]
blue = rgb[:, :, 2]
```

```python
plt.title("Histogram")
plt.hist(red.ravel(), bins=256, color="r", label="Red")
plt.hist(green.ravel(), bins=256, color="g", label="Green")
plt.hist(blue.ravel(), bins=256, color="b", label="Blue")
plt.xlabel('Intensity')
plt.ylabel('Counts')
plt.legend()
plt.show()
```
